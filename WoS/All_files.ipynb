```python
def trigramas(words):
 
    trigrams = []
 
    for i in range(0, len(words)):
            try:
                trigrama_obs = words[i] + ' ' + words[i+1] + ' ' + words[i+2]
                trigrams.append(trigrama_obs)
            except:
                pass
    return trigrams

```
